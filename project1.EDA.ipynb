```python
plt.figure(figsize=(12,5))
sns.histplot(p['price'],bins=30,kde=True)
plt.title('Histogram of Price')
plt.vlines(p['price'].mean(),ymin=0,ymax=130,colors='blue',label='Mean')
plt.vlines(p['price'].median(),ymin=0,ymax=130,colors='red',label='Median')
plt.vlines(p['price'].mode(),ymin=0,ymax=130,colors='c',label='Mode')
plt.legend()
plt.xticks(np.arange(1000000,13500000,1000000))
plt.yticks(np.arange(0,150,20))
plt.show()

```
